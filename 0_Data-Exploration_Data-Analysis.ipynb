# Check Missing Values:

```python
m_values = t_data.isnull().sum()
m_values[m_values > 0]
```

```python
#  Filling numerical missing_values with median:
numerical_cols = t_data.select_dtypes(include=['float64', 'int64']).columns
t_data[numerical_cols] = t_data[numerical_cols].apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
# Filling categorical missing_values with 'Unknown'
categorical_cols = t_data.select_dtypes(include=['object']).columns
t_data[categorical_cols] = t_data[categorical_cols].apply(lambda x: x.fillna('Unknown'))
```

```python
t_data['Start'] = t_data['Start'].fillna('2019-04-29 07:08:38')
```

```python
t_data['End'] = t_data['End'].fillna('2019-04-25 00:01:32')
```

```python
t_data.isnull().sum().iloc[0:55]
```

# Outliers by Quantile method:

```python
Q1 = t_data[numerical_cols].quantile(0.25)
Q3 = t_data[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = t_data[(t_data[numerical_cols] < lower_bound) | (t_data[numerical_cols] > upper_bound)]
```

# Box_plot for Outliers:

```python
plt.figure(figsize=(10,4),dpi=200)
outliers.plot(kind = 'box',color = 'blue')
plt.xticks(rotation = 90)
plt.title('Total outliers', color = 'blue')
plt.ylabel('Values',color = 'blue')
plt.show()
```
